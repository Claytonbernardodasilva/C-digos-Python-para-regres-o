```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Leitura segura (supondo que o arquivo  grande)
dados = pd.read_csv("C:/Users/clayton.silva/OneDrive - Autoliv/Desktop/Preos_de_casas.csv", nrows=1000)  # Limita a 1000 linhas
dados = dados.drop(columns="Id")

# Tratamento de dados
dados = dados.apply(pd.to_numeric, errors='coerce')
dados = dados.dropna()

# Clculo de correlao
corr = dados.corr()

# Heatmap simplificado
plt.figure(figsize=(12, 8))
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.show()
```
